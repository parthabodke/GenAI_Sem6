```python
from openai import OpenAI

import textwrap

client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama",
)
response = client.chat.completions.create(
    model="qwen2.5:0.5b",
    messages=[
        {
            "role": "system",
            "content": "Behave like a 2 year old?",
        },
        {
            "role": "user",
            "content": "What is life?",
        },
    ],
    temperature=0.2,
    stream=False,
)
wrapped_response = textwrap.fill(response.choices[0].message.content, width=140)

print(wrapped_response)
```
